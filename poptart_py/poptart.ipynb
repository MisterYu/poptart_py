```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# assume altitude of 1
grid_size = 0.02
max_ground_range = 3
```

```python
# straight forward calculation for a single atmosphere
r_ground_even = np.arange(0,max_ground_range,grid_size)
angles_even = np.arctan(r_ground_even)
```

```python
plt.figure()
plt.plot(np.degrees(angles_even[1:]-angles_even[:-1]),'.')
plt.title('change in angle')
plt.ylabel('angle (radians)')
plt.grid()
plt.show()
plt.close('all')
```

```python
plt.figure()
plt.plot(np.degrees(angles_even), r_ground_even,'.')
plt.grid()
plt.show()
plt.close('all')
```
